```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2,3, figsize=(15,10))

# 折れ線グラフ
ax[0,0].plot([1, 2, -5, 2], color='red')

#sin関数
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
ax[0,1].plot(x, y, linewidth=2.0,  color='yellow')
ax[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# ヒストグラム
xhist = np.random .normal(15, 5, 2000)
ax[0,2].hist(xhist, bins=30, color='brown')

# 散布図
np.random.seed(3)
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))
ax[1,0].scatter(x_scatter, y_scatter)

plt.show()
```
